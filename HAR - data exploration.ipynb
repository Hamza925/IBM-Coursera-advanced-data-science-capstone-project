```python
import pandas as pd
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X_train = pd.read_csv('Downloads/Datasets/UCI HAR Dataset/UCI HAR Dataset/train/X_train.txt', header=None, delim_whitespace=True)
```

```python
%matplotlib inline
```

```python
y_train = pd.read_csv('Downloads/Datasets/UCI HAR Dataset/UCI HAR Dataset/train/y_train.txt', delim_whitespace=True, header=None, names = ['Activity'])
y_train.Activity.replace([1, 2, 3, 4, 5, 6], ['WALKING',
'WALKING_UPSTAIRS',
'WALKING_DOWNSTAIRS',
'SITTING',
'STANDING',
'LAYING'
], inplace =True)
```

now we will plot the distribution of data for each category of activity..
this is very basic and important later in evaluation of our model. 

```python
activity_count  = y_train['Activity'].value_counts()
plt.figure(figsize=(15,5))
sns.barplot(activity_count.index, activity_count.values, alpha=0.8)
plt.title('Activity Histogram')
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('Activity Type', fontsize=12)
plt.show()
```
